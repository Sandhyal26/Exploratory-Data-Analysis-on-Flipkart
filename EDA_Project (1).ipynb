##### 8. Clean the Data : 

```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("laptop_data.csv") 
print(" Data loaded successfully")
print(df.head())
print(df.info())
```

```python
# Strip extra spaces and fix casing in headers
```

```python
df.columns = df.columns.str.strip().str.title().str.replace(" ", "_")

print("Cleaned Headers:")
print(df.columns.tolist())
```

```python
# extract numeric values from Display or combine details if needed.
```

```python
# Extract numeric display size from Display column
df['Display_Inch'] = df['Display'].str.extract(r'\(([\d\.]+)\s*Inch\)').astype(float)

# Split Processor into Brand and Model
df['Processor_Brand'] = df['Processor'].str.split().str[0]
df['Processor_Model'] = df['Processor'].str.extract(r'([Rr]yzen\s*\d|[Cc]ore\s*[iI]\d|Celeron|Snapdragon|Kompanio)')

print("Columns split and formatted.")
print(df[['Processor', 'Processor_Brand', 'Processor_Model', 'Display_Inch']].head())
```

```python
# Convert numeric-like columns to numeric
num_cols = ['Display_Inch', 'Ratings', 'Gross_Income']
for col in num_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert Windows type to categorical
df['Windows'] = df['Windows'].astype('category')

print(" Data types converted successfully.")
print(df.dtypes)
```

```python
# Check for missing values
print("Missing Values:")
print(df.isnull().sum())

# Fill or drop missing values
df['Rating'].fillna(df['Rating'].median())
df['Display_Inch'].fillna(df['Display_Inch'].median())

# Optional: drop rows if many values missing
df.dropna(subset=['Processor', 'Windows'])

print("Missing values handled.")
```

```python
#the IQR (Interquartile Range) method to detect outliers in numeric data.
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_limit) | (df[col] > upper_limit)]
    print(f"{col}  Outliers detected: {len(outliers)}")
    

    
print(" Outliers capped using IQR method.")
```

```python
#Merge Processor Brand and Model into one descriptive column
df['Processor_Details'] = df['Processor_Brand'] + " " + df['Processor_Model'].fillna('')

#Split Windows version and type
df[['Windows_Version', 'Windows_Type']] = df['Windows'].astype(str).str.extract(r'(Windows\s*\d+)\s*(\w+)?')

print("Columns merged/split successfully.")
print(df[['Processor_Details', 'Windows_Version', 'Windows_Type']].head())
```

```python
print("Final Data Overview:")
print(df.info())
print(df.describe(include='all'))
```

```python
df.to_csv("cleaned_laptop_data.csv", index=False)
```
